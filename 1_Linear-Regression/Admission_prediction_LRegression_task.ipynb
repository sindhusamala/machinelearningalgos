```python

import pandas as pd
import numpy as np 
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
data= pd.read_excel("Admission_Prediction.xlsx")
data.head()
```

```python
data.drop('Serial No.', axis=1, inplace=True)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['GRE Score'] = data['GRE Score'].replace( np.NaN, data['GRE Score'].mean())
data['TOEFL Score'] = data['TOEFL Score'].replace( np.NaN, data['TOEFL Score'].mean())
data['University Rating'] = data['University Rating'].replace( np.NaN, data['University Rating'].mean())
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
cor= data.corr()
sns.heatmap(cor)
```

```python
X= data.loc[:,['GRE Score','TOEFL Score','University Rating', 'SOP','LOR','CGPA','Research']]
Y= data.loc[:,['Chance of Admit']]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X,Y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
data.head()
```

```python
predval= model.predict([[330,120,3.0,4.0,4.5,8.90,1.0]]) 
predval
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=10, test_size=0.3)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_p = model.predict(X_train)
Y_p
```

```python
Y_predict = model.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
R2Score =  r2_score(Y_test,Y_predict)
print(R2Score)
print("MSE: ", mean_squared_error(Y_test, Y_predict))
print("MAE: ", mean_absolute_error(Y_test, Y_predict))
```

```python
print("R2Score for training: ", r2_score(Y_train,Y_p))
print("MSE for training: ", mean_squared_error(Y_train,Y_p))
print("MAE for training: ", mean_absolute_error(Y_train,Y_p))
```
