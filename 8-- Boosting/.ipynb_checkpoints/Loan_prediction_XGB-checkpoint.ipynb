### Handling Outliers

### 1. ApplicantIncome--- The data is normally distributed so we use 3 sigma rule 

```python
from scipy import stats
```

```python
lower_limit = new_train.ApplicantIncome.mean() - 3* new_train.ApplicantIncome.std()
print(lower_limit)

upper_limit = new_train.ApplicantIncome.mean() + 3* new_train.ApplicantIncome.std()
upper_limit
```

```python
new_train.sort_values("ApplicantIncome")
```

```python
new_train.loc[new_train["ApplicantIncome"]<lower_limit]
```

```python
new_train.loc[new_train["ApplicantIncome"]>upper_limit]
```

```python
# Imputing outliers with mean as it is normal distribution
new_train.loc[new_train["ApplicantIncome"]>upper_limit, 'ApplicantIncome']= np.mean(new_train.ApplicantIncome)
```

```python
new_train.loc[new_train["ApplicantIncome"]>upper_limit]
```

### 2. CoapplicantIncome

```python
lower_limit = new_train.CoapplicantIncome.mean() - 3* new_train.CoapplicantIncome.std()
print(lower_limit)

upper_limit = new_train.CoapplicantIncome.mean() + 3* new_train.CoapplicantIncome.std()
upper_limit
```

```python
new_train.loc[new_train["CoapplicantIncome"]<lower_limit]
```

```python
new_train.loc[new_train["CoapplicantIncome"]>upper_limit]
```

```python
new_train.loc[new_train["CoapplicantIncome"]>upper_limit, 'CoapplicantIncome'] = np.mean(new_train.CoapplicantIncome)
```

```python
new_train.loc[new_train["CoapplicantIncome"]>upper_limit]
```

### 3. LoanAmount-- there is skweness in data so we replace outliers with its median

```python
##using IQR to remove the outliers
IQR = stats.iqr(new_train.LoanAmount,interpolation = 'midpoint') 

IQR
```

```python
Q1 = new_train.LoanAmount.quantile(0.25)
 
Q3 = new_train.LoanAmount.quantile(0.75)

min_limit=Q1 - 1.5*IQR

max_limit=Q3 +1.5*IQR
```

```python
min_limit
```

```python
max_limit
```

```python
new_train.loc[new_train['LoanAmount'] < min_limit]
```

```python
new_train.loc[new_train['LoanAmount'] > max_limit]
```

```python
new_train.loc[new_train['LoanAmount'] > max_limit,'LoanAmount']=np.median(new_train.LoanAmount)
```

```python
new_train.loc[new_train['LoanAmount'] > max_limit]
```

## Model Creation

```python
new_train.head()
```

```python
X = new_train.iloc[:,1:-1]
Y = new_train.Loan_Status
```

```python
new_train.Loan_Status.value_counts()
```

```python
X
```

```python
Y
```

### Balencing Loan_Status data

```python
sns.countplot(Y)
```

```python
# Balencing the data

from imblearn.over_sampling import SMOTE
smote= SMOTE()
```

```python
X_smote, Y_smote= smote.fit_resample(X,Y)
```

```python
from collections import Counter
print('Actual classes', Counter(Y))
print('Smote classes',Counter(Y_smote))
```

```python
sns.countplot(Y_smote);
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X_smote,Y_smote, test_size= 0.25,random_state= 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

### We will implement four classification algorithms,

1. Logistic Regression Classifier

2. Decision Trees Classifier

3. Random Forest Classifier

4. K Nearest Neighbours Classifier

### 1. Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()       
log_reg.fit(X_train,Y_train)   
```

```python
Y_train_pre=log_reg.predict(X_train)
Y_pred = log_reg.predict(X_test)
Y_pred
```

```python
Y_train.shape
```

```python
Y_pred.shape
```

```python
## Accuracy for training data
from sklearn.metrics import accuracy_score,auc, confusion_matrix,precision_score,recall_score,f1_score,classification_report
accuracy = accuracy_score(Y_train,Y_train_pre)
accuracy  
```

```python
# model test accuracy  
accuracy = accuracy_score(Y_test,Y_pred)
accuracy
```

```python
Precision = precision_score(Y_test,Y_pred)
Precision
```

```python
report=classification_report(Y_test, Y_pred)
print(report)
```

```python
## checking cross validation score
from sklearn.model_selection import cross_val_score

scores = cross_val_score(log_reg,X,Y,cv=3,scoring='f1')
print(scores)
print("Cross validation Score:",scores.mean()) #get the mean value of scores to determine the overall accuracy of the model
print("Std :",scores.std())
#std of < 0.05 is good 
```

```python
from sklearn.model_selection import GridSearchCV
  
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'penalty' : ['l1', 'l2'], 
              'max_iter': list(range(100,800,100)),
              'solver': ['newton-cg', 'Ibfgs', 'Liblinear', 'sag','saga']}
             
grid = GridSearchCV(log_reg, param_grid, refit = True, verbose = 3,scoring='f1',cv=3)
  
# fitting the model for grid search
grid.fit(X_train, Y_train)
grid.best_params_

#Summerise
print("mean Accuracy: %.3f", grid.best_score_)
print("Config: %s", grid.best_params_)
```

```python
# print best parameter after tuning
print(grid.best_params_)
  
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```
